```python
with open("prompts.json", "r") as f:
    data = json.load(f)

lengths = []
for item in data:
    text = item["prompt"]
    ids = tokenizer.encode(text)
    lengths.append(len(ids))

print("Token counts:", lengths)
print("Min:", np.min(lengths))
print("Max:", np.max(lengths))
print("Mean:", np.mean(lengths))
print("Median:", np.median(lengths))

# Optional: print the longest example
max_index = np.argmax(lengths)
print("\nLongest sample:\n", data[max_index]["prompt"])
print("\nLength:", lengths[max_index])
```

```python
from datasets import load_dataset

dataset = load_dataset("json", data_files="prompts.json")
dataset
```

```python
from transformers import AutoTokenizer

model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
```

```python
def tokenize_fn(batch):
    enc = tokenizer(
        batch["prompt"],
        padding="max_length",
        truncation=True,
        max_length=128,
        return_tensors=None
    )
    # Labels are just input_ids for causal LM
    enc["labels"] = enc["input_ids"].copy()
    return enc

tokenized = dataset.map(tokenize_fn, batched=True)
```

```python
from transformers import AutoModelForCausalLM, TrainingArguments, Trainer

model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
args = TrainingArguments(
    output_dir="./gpt2-finetuned-demo",
    per_device_train_batch_size=2,
    learning_rate=5e-5,
    num_train_epochs=10,
    weight_decay=0.01,
    logging_steps=1,
    save_steps=20,
    fp16=False,   # You can switch to fp16=True if GPU supports it
)
```

```python
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized["train"],
    eval_dataset=None,
)
```

```python
trainer.train()
```

```python
text = "### Instruction:\nRewrite politely: 'Send me the files now!'\n\n### Response:\n"

inputs = tokenizer.encode(text, return_tensors="pt")
outputs = model.generate(
    inputs,
    max_length=80,
    temperature=0.7,
    do_sample=True
)
print(tokenizer.decode(outputs[0]))
```
